```python
import plotly.graph_objects as go
import pandas as pd

# Load the data from a CSV file
df = pd.read_csv('/home/long/Downloads/test.csv')

# Extract the data you want to plot
data = df['FurLength'].values
data1 = df['Type'].values

FurLength = []
countcat = 0
countdog = 0
for cat in data1:
    if cat == 2:
        countcat +=1
for dog in data1:
    if dog == 1:
        countdog += 1
count0 = 0
count1 = 0
count2 = 0
count3 = 0
for t0 in data:
    if t0 == 2:
        count0 +=1

for t1 in data:
    if t1 == 0:
        count1 +=1
for t2 in data:
    if t2 == 2:
        count2 +=1
for t3 in data:
    if t3 == 3:
        count3 +=1
# print(count2)
FurLength.append(count0)
FurLength.append(count1)
FurLength.append(count2)
FurLength.append(count3)
print(FurLength)
# Create a histogram trace with two values
trace0 = go.Histogram(
    x=[count0],
    name='Short',
    xbins=dict(
        start=min(FurLength) - 50,
        end=max(FurLength) + 50,
        size=100
    ),
    opacity=0.75,
    marker=dict(color='Blue')
)
trace0.name='Short'
trace1 = go.Histogram(
    x=[count1],
    name='Medium',
    xbins=dict(
        start=min(FurLength) - 50,
        end=max(FurLength) + 50,
        size=100
    ),
    opacity=0.75,
    marker=dict(color='red')
)
trace1.name='Medium'

trace2 = go.Histogram(
    x=[count2],
    name='Long',
    xbins=dict(
        start=min(FurLength) - 50,
        end=max(FurLength) + 50,
        size=100
    ),
    opacity=0.75,
    marker=dict(color='Green')
)
trace2.name='Long'

trace3 = go.Histogram(
    x=[count3],
    name='Not Specified',
    xbins=dict(
        start=min(FurLength) - 50,
        end=max(FurLength) + 50,
        size=100
    ),
    opacity=0.75,
    marker=dict(color='Black')
)
trace3.name='Not Specified'
# Set the layout
layout = go.Layout(
    title='Compare which animals are abandoned more',
    xaxis=dict(title='Quantity'),
    # yaxis=dict(title=''),
    bargap=0.05,
    bargroupgap=0.2,
    width=1000,
    height=800
)

# Create the figure object and plot it
fig = go.Figure(data=[trace0,trace1,trace2,trace3], layout=layout)
fig.show(renderer = "png")
```
