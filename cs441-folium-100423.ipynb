```python
import pandas as pd
import folium
location = "/home/long/Downloads/dublin.csv"
bike_station_locations = pd.read_csv(location)
bike_station_locations_10 = bike_station_locations[:11]
bike_station_locations_data = bike_station_locations_10[["Latitude", "Longitude", "Name"]]
# print(bike_station_locations_10)
```

```python
map = folium.Map(location=[bike_station_locations_data.Latitude.mean(), bike_station_locations_data.Longitude.mean()], zoom_start=14, control_scale=True)
for index, location_info in bike_station_locations_data.iterrows():
    folium.Marker([location_info["Latitude"], location_info["Longitude"]], popup=location_info["Name"]).add_to(map)
map
```
