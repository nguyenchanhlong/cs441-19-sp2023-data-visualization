```python
import pandas as pd
import matplotlib.pyplot as plt

# read csv file and this is the main data for draw barchart
test_data = pd.read_csv('/home/long/Downloads/test.csv')

# Define labels for fur lengths
labels = ['Short', 'Medium', 'Long', 'Not Specified']

# Define a function to count the number of animals with a certain fur length

def count_animals_with_fur_length(animal_type, fur_length):
    return len(test_data[(test_data['Type'] == animal_type) & (test_data['FurLength'] == fur_length)])

# Count the number of dogs and cats with each fur length
total_fur_dogs = [count_animals_with_fur_length(1, i) for i in range(4)]
total_fur_cats = [count_animals_with_fur_length(2, i) for i in range(4)]

# Plot the barchart for dogs
plt.bar(labels, total_fur_dogs, color='#3498db', alpha=0.7)
plt.xlabel('Type of Fur')
plt.ylabel('Quantity')
plt.title('What are the most abandoned DOGS with fur length?')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Plot the barchart for cats
plt.bar(labels, total_fur_cats, color='#3498db', alpha=0.7)
plt.xlabel('Type of Fur')
plt.ylabel('Quantity')
plt.title('What are the most abandoned CATS with fur length?')
plt.grid(axis='y', alpha=0.75)
plt.show()

```
