```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

train = pd.read_csv('/home/long/Downloads/train.csv')
test = pd.read_csv('/home/long/Downloads/test.csv')
all_data = pd.concat([train, test])
```

```python
fig, ax = plt.subplots(figsize=(16, 12))
try:
    plt.subplot(1, 2, 1)
    text_cat = ' '.join(all_data.loc[all_data['Type'] == 2, 'Name'].fillna('').values)
    if not text_cat:
        raise ValueError("No text data found for cats.")
    wordcloud = WordCloud(max_font_size=None, background_color='white',
                          width=1200, height=1000).generate(text_cat)
    plt.imshow(wordcloud)
    plt.title('Top cat names')
    plt.axis("off")
    
    
    plt.subplot(1, 2, 2)
    text_dog = ' '.join(all_data.loc[all_data['Type'] == 1, 'Name'].fillna('').values)
    if not text_dog:
        raise ValueError("No text data found for dogs.")
    wordcloud = WordCloud(max_font_size=None, background_color='white',
                          width=1200, height=1000).generate(text_dog)
    plt.imshow(wordcloud)
    plt.title('Top dog names')
    plt.axis("off")
    
    
    plt.show()
    
    
except ValueError as e:
    print("ValueError: ", e)
```
