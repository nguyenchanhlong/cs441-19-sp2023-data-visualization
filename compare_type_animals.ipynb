```python
import plotly.graph_objects as go
import pandas as pd

# Load the data from a CSV file
df = pd.read_csv('/home/long/Downloads/test.csv')

# Extract the data you want to plot
data = df['Type'].values
count2 = 0
count1 = 0
for t1 in data:
    if t1 == 1:
        count1 +=1
for t2 in data:
    if t2 == 2:
        count2 +=1
print(count2)

# Create a histogram trace with two values
trace1 = go.Histogram(
    x=[count1],
    name='Cat',
    xbins=dict(
        start=min(count1, count2) - 5,
        end=max(count1, count2) + 5,
        size=20
    ),
    text='Dog',
    opacity=0.75,
    marker=dict(color='Blue')
)
# trace1.name='Cat'

trace2 = go.Histogram(
    x=[count2],
    name='Dog',
    xbins=dict(
        start=min(count1, count2) - 5,
        end=max(count1, count2) + 5,
        size=20
    ),
    text='Cat',
    opacity=0.75,
    marker=dict(color='red')
)
# trace2.name='Dog'
# Set the layout
layout = go.Layout(
    title='Compare which animals are abandoned more',
    xaxis=dict(title='Quantity'),
    # yaxis=dict(title=''),
    bargap=0.05,
    bargroupgap=0.3,
    width=800,
    height=600
)

# Create the figure object and plot it
fig = go.Figure(data=[trace1,trace2], layout=layout)
fig.show(renderer = 'png')
```
